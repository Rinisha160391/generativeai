```python
import os
os.environ['SERPAPI_API_KEY']="**"
os.environ['OPENAI_API_KEY']="**"
#os.environ[]=
```

```python
from langchain.utilities import SerpAPIWrapper
from langchain.agents import Tool
from langchain.tools.file_management.write import WriteFileTool
from langchain.tools.file_management.read import ReadFileTool

search = SerpAPIWrapper()
tools = [
    Tool(
        name = "search",
        func=search.run,
        description="useful to know about the current events as chat models knowledge are limited to specific timelines"
    ),
    WriteFileTool(),
    ReadFileTool(),
]
```

```python
from langchain.vectorstores import FAISS
from langchain.docstore import InMemoryDocstore
from langchain.embeddings import OpenAIEmbeddings
import faiss
```

```python
embeddings_model=OpenAIEmbeddings()
index=faiss.IndexFlatL2(1536)#OpenAI embeddingsize 1536
vectorstore=FAISS(embeddings_model.embed_query, index, InMemoryDocstore({}), {})
```

```python
from langchain_experimental.autonomous_agents import AutoGPT
from langchain.chat_models import ChatOpenAI
```

```python
agent = AutoGPT.from_llm_and_tools(
    ai_name="Rinisha",
    ai_role="Assistant",
    tools=tools,
    llm=ChatOpenAI(temperature=0),
    memory=vectorstore.as_retriever()
)
# Set verbose to be true
agent.chain.verbose = True
```

```python
agent.run("Recommend 5 movies to watch which is similar to Avengers")
```
